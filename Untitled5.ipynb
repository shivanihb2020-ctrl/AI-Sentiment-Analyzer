```python
from transformers import pipeline
# we are creating a classifier that specializes in sentiment
classifier=pipeline("sentiment-analysis")
```

```python
sentence="I am so frustrated becausev this code is taking forever to run!"
result=classifier(sentence)
print(result)
```

```python
sentence="I am so happy today"
result=classifier(sentence)
print(result)
```

```python
import ipywidgets as widgets
from IPython.display import display

# 1. Create the Text Box
input_text = widgets.Text(
    placeholder='Type a sentence here...',
    description='Sentence:',
    style={'description_width': 'initial'}
)

# 2. Create the "Analyze" Button
button = widgets.Button(description="Analyze Sentiment", button_style='success')
output = widgets.Output()

# 3. Define what happens when you click the button
def on_button_clicked(b):
    with output:
        output.clear_output()
        if input_text.value:
            # This uses the 'classifier' we built in Step 3
            result = classifier(input_text.value)
            label = result[0]['label']
            score = round(result[0]['score'] * 100, 2)
            print(f"Prediction: {label} ({score}% confidence)")
        else:
            print("Please type something first!")

button.on_click(on_button_clicked)

# 4. Display the "App"
display(input_text, button, output)
```

```python
import matplotlib.pyplot as plt

def show_sentiment_chart(label, score):
    # Create a simple bar chart
    fig, ax = plt.subplots(figsize=(6, 1))

    # Set color: Green for positive, Red for negative
    color = 'green' if label == 'POSITIVE' else 'red'

    # Draw the bar (score is a percentage)
    ax.barh(['Sentiment'], [score], color=color)
    ax.set_xlim(0, 100)
    ax.set_title(f"AI is {score}% sure this is {label}")

    plt.show()

# Let's test the chart function with a fake score
show_sentiment_chart("POSITIVE", 95)
```

```python
output = widgets.Output()

def on_button_clicked_v2(b):
    with output:
        output.clear_output()
        text = input_text.value
        if text:
            # 1. Get AI Result
            result = classifier(text)[0]
            label = result['label']
            score = round(result['score'] * 100, 2)

            # 2. Show the Chart
            show_sentiment_chart(label, score)
        else:
            print("Type something first!")

button.on_click(on_button_clicked_v2)
display(input_text, button, output)
```
